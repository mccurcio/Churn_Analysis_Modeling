## EDA Pie Chart

- **Accepted all columns as unchanged.**
- Only columns 1 & 17 have skew >= 2.0.
- This is ok since they are binary


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import scipy.stats as stats

# Load Cleaned Churn Data
df = pd.read_csv("../data/processed/mcc_clean_churn.csv")
```

```python
labels = 'Exited', 'Retained'
sizes = [df.Churned[df['Churned']==1].count(), df.Churned[df['Churned']==0].count()]
explode = (0, 0.1)
fig1, ax1 = plt.subplots(figsize=(10, 8))
ax1.pie(sizes, 
        explode=explode, 
        labels=labels, 
        autopct='%1.0f%%',
        shadow=True, 
        startangle=90)
ax1.axis('equal')
plt.title("Proportion of customer churned and retained, ~1:7.", size = 20)
plt.show()
```
